```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

### Let's define some transformations to resize the image and put it under tensor format.

```python

transform = transforms.Compose([
    transforms.Resize((64, 64)),  
    transforms.ToTensor()         
])
```

### Then let's download the dataset under "data" folder

```python
dataset = datasets.CelebA(root='./data', split='train', download=True, transform=transform)
```

```python
print(f"Total downloaded images : {len(dataset)}")
```

### Display some of the images

```python
images = [dataset[i][0] for i in range(16)]

fig, axes = plt.subplots(4, 4, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i].permute(1, 2, 0))  # Convert from CxHxW to HxWxC
    ax.axis('off')
plt.show()
```
